```python
import pandas as pd
import scanpy as sc

# Assuming that single_cell3 is your AnnData object and it has been properly loaded

# Make sure that the variable names (which should be gene names) are set correctly
exp_df = single_cell3.to_df()
exp_df.columns = single_cell3.var['feature_name'].values

# Now group by cell type and calculate the mean expression
grouped_exp = exp_df.groupby(single_cell3.obs['cell_type']).mean()

# This DataFrame should now have gene names as columns
print(grouped_exp)
```

```python
import pandas as pd
from scipy.sparse import issparse

# Check if the expression matrix is sparse and convert it to a dense format if necessary
if issparse(single_cell3.X):
    matrix_dense = single_cell3.X.toarray()
else:
    matrix_dense = single_cell3.X

# Convert the dense matrix to a DataFrame for easier manipulation
exp_df = pd.DataFrame(matrix_dense, columns=single_cell3.var_names, index=single_cell3.obs_names)

# Convert the expression data to binary: 1 if expression > 0, else 0
binary_exp_df = exp_df.gt(0).astype(int)

# Group by cell type and calculate the fraction of cells expressing each gene
fraction_cells_exp = binary_exp_df.groupby(single_cell3.obs['cell_type']).mean()

# The result is a DataFrame where each row represents a cell type,
# and each column represents a gene. The values are the fractions of cells
# expressing each gene within each cell type.
print(fraction_cells_exp)
```
